```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 4 * np.pi, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 2, 1)
ax.plot(x, y)
ax.set_xlabel("X")
ax.set_ylabel("$\sin(x)$")

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(y, x)
ax2.set_xlabel("$\sin(x)$")
ax2.set_ylabel("X")
```
