## Use np.random.randn to plot a normal distribution, using random samples

```python
x = np.random.randn(1000)
```

```python
plt.hist(x)
```

```python
samples = np.random.randn(1000000)
```

```python
plt.hist(samples, bins=100)
```

```python
plt.hist?
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)

counts, bins, _ = ax1.hist(samples, bins=100)

cdf = counts.cumsum()

ax2 = fig.add_subplot(2, 1, 2)

centers = (bins[:-1] + bins[1:])/2
ax2.plot(centers, cdf)
```

## Data Exploration

```python
import pandas as pd
```

```python
df = pd.read_csv('data/winequality-white.csv', delimiter=';')
```

```python
ax = df.plot.scatter('alcohol', 'density')
ax.set_xlabel('Alcohol Content')
ax.set_xticks([])
ax.figure.suptitle('Density vs. Alcohol Content')
```
