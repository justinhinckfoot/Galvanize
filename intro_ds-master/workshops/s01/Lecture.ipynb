## FFT

```python
x = np.arange(0, 4 * np.pi, .1)
y = np.sin(x) + np.sin(2*x) + np.sin(3*x)
```

```python
plt.plot(x/np.pi, y)
```

```python
plt.plot(fft(y)[:len(y)//2])
```

```python
plt.plot(ifft(fft(y)))
```

## Optimization

```python
from scipy.optimize import minimize
```

```python
def smooth_function(inputs):
    return ((inputs - 4)**2).sum()
```

```python
minimize(smooth_function, x0=np.array([2, 9, 8]))
```

```python
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(42)

x = np.linspace(0, 40, 1000)

y = np.sin(np.linspace(0, 10*np.pi, 1000))
y += np.random.randn(len(x))

fig = plt.figure()

ax = fig.add_subplot(1,1,1)

ax.plot(x, y)
```

```python
from matplotlib import gridspec

fig = plt.figure(figsize=(7, 3))
gs = gridspec.GridSpec(1, 2, width_ratios=[3,1])

# plot the timeseries
ax0 = plt.subplot(gs[0])
ax0.plot(x, y)

# let's add some axes labels to boot
ax0.set_ylabel(r'displacement ($\AA$)')
ax0.set_xlabel('time (ns)')

# and now we'll also refine the y-axis ticks a bit
ax0.set_ylim(-4.5, 4.5)

# plot the distribution
ax1 = plt.subplot(gs[1])
ax1.hist(y, histtype='step', bins=40, normed=True, orientation='horizontal')

# this will remove the grid and the ticks on the x-axis
ax1.set_xticks([])
ax1.grid(False)

ax1.set_ylim(-4.5, 4.5)
```

### 3D plotting

```python
time = np.linspace(0, 10*np.pi, 1000)

pos = np.vstack([np.sin(2*time),
                 np.sin(time),
                 np.sin(time)])
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x, y, z = pos
ax.plot(x, y, z)
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
ax.set_zlabel(r"$z$")
```
